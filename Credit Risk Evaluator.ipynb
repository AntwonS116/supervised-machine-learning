```python
import numpy as np
import pandas as pd
from pathlib import Path
from sklearn.preprocessing import LabelEncoder
```

```python
train_df = pd.read_csv(Path('Resources/2019loans.csv'))
test_df = pd.read_csv(Path('Resources/2020Q1loans.csv'))
```

```python
train_df.head()
```

```python
test_df.head()
```

```python
# Convert categorical data to numeric and separate target feature for training data
x_train = train_df.drop("loan_status", axis=1)
y_train = train_df["loan_status"]
```

```python
#Label Encoding 
y_train = LabelEncoder().fit_transform(train_df["loan_status"])
y_train
```

```python
# add missing dummy variables to testing set
train_features = pd.get_dummies(x_train)
print(train_features.columns)
```

```python
# Convert categorical data to numeric and separate target feature for testing data
x_test = test_df.drop("loan_status", axis=1)
y_test = test_df["loan_status"]
```

```python
# add missing dummy variables to testing set
test_features = pd.get_dummies(x_test)
print(test_features.columns)
```

```python
#Label Encoding
y_test = LabelEncoder().fit_transform(test_df["loan_status"])
y_test
```

```python
x_train
```

```python
for column in train_features.columns:
    if column not in test_features.columns:
        test_features[column] = 0
```

### Initial Thought: 
After doing some research on what the random forest classifier is most commonly used for, I believe it is better for categorical data. I think this model will perform better than the logistic regression model.

```python
# Train the Logistic Regression model on the unscaled data and print the model score

from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(solver='lbfgs', max_iter=30000)
classifier.fit(train_features, y_train)
print(f"Training Data Score: {classifier.score(train_features, y_train)}")
print(f"Testing Data Score: {classifier.score(test_features, y_test)}")
```

```python
# Train a Random Forest Classifier model and print the model score

from sklearn.ensemble import RandomForestClassifier
f_classifier = RandomForestClassifier(random_state=1, n_estimators=500).fit(train_features, y_train)
print(f"Training Data Score: {f_classifier.score(train_features, y_train)}")
print(f"Testing Data Score: {f_classifier.score(test_features, y_test)}")
```

### Before Scaling 
The Random Forest Classifier seems to have better scores before scaling. However, with a 1.0 score on the training data set it raises concerns of over fitting.

```python
# Scale the data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler().fit(train_features)
train_features_scaled = scaler.transform(train_features)
test_features_scaled = scaler.transform(test_features)
```

```python
# Train the Logistic Regression model on the scaled data and print the model score
classifier.fit(train_features_scaled, y_train)
print(f"Training Data Score: {classifier.score(train_features_scaled, y_train)}")
print(f"Testing Data Score: {classifier.score(test_features_scaled, y_test)}")
```

```python
# Train a Random Forest Classifier model on the scaled data and print the model score
from sklearn.ensemble import RandomForestClassifier
f_classifier = RandomForestClassifier(random_state=1, n_estimators=500).fit(train_features_scaled, y_train)
print(f"Training Data Score: {f_classifier.score(train_features_scaled, y_train)}")
print(f"Testing Data Score: {f_classifier.score(test_features_scaled, y_test)}")
```

### After Scaling / Conclusions 
The logistic regression model has better scores than the random forest classifier. This is contrary to my initial thoughts. The training data set with random forest is still sitting at a 1.0 score and this is a sign of over fitting.
